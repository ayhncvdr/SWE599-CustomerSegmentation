## Clustering using K- means
### 1.Segmentation using Age and Spending Score

```python
'''Age and spending Score'''
X1 = df[['Age' , 'Spending Score (1-100)']].iloc[: , :].values
inertia = []
for n in range(1 , 11):
    algorithm = (KMeans(n_clusters = n ,init='k-means++', n_init = 10 ,max_iter=300, 
                        tol=0.0001,  random_state= 111  , algorithm='elkan') )
    algorithm.fit(X1)
    inertia.append(algorithm.inertia_)
```

```python

silhouette_scores = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, init='k-means++', n_init=10, max_iter=300, random_state=111)
    kmeans.fit(X1)
    silhouette_scores.append(silhouette_score(X1, kmeans.labels_))

# Plot silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Score for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()


db_scores = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, init='k-means++', n_init=10, max_iter=300, random_state=111)
    kmeans.fit(X1)
    db_scores.append(davies_bouldin_score(X1, kmeans.labels_))

# Plot Davies-Bouldin scores
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), db_scores, marker='o')
plt.title('Davies-Bouldin Index for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('Davies-Bouldin Index')
plt.show()
```

```python
# 4 is chosen as optimal cluster number
algorithm = (KMeans(n_clusters = 4,init='k-means++', n_init = 10 ,max_iter=300, 
                        tol=0.0001,  random_state= 111  , algorithm='elkan') )
algorithm.fit(X1)
labels1 = algorithm.labels_
centroids1 = algorithm.cluster_centers_
```

```python
h = 0.02
x_min, x_max = X1[:, 0].min() - 1, X1[:, 0].max() + 1
y_min, y_max = X1[:, 1].min() - 1, X1[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = algorithm.predict(np.c_[xx.ravel(), yy.ravel()]) 
```

```python
plt.figure(1 , figsize = (15 , 7) )
plt.clf()
Z = Z.reshape(xx.shape)
plt.imshow(Z , interpolation='nearest', 
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap = plt.cm.Pastel2, aspect = 'auto', origin='lower')

plt.scatter( x = 'Age' ,y = 'Spending Score (1-100)' , data = df , c = labels1 , 
            s = 200 )
plt.scatter(x = centroids1[: , 0] , y =  centroids1[: , 1] , s = 300 , c = 'red' , alpha = 0.5)
plt.ylabel('Spending Score (1-100)') , plt.xlabel('Age')
plt.show()
```

## Clustering using Hierarchical Clustering (Agglomerative Clustering)
### 1.Segmentation using Age and Spending Score

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch

# Hierarchical clustering for 'Age' and 'Spending Score (1-100)'
X1 = df[['Age' , 'Spending Score (1-100)']].values
```

```python
# Using the dendrogram to find the optimal number of clusters
dendrogram = sch.dendrogram(sch.linkage(X1, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.axhline(y=300, c='r', linestyle='dashed')
plt.show()
```

```python
# Fitting hierarchical clustering to the dataset
# 2 is chosen as optimal cluster number
hc = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward') 
y_hc = hc.fit_predict(X1)
```

## Clustering using DBSCAN Clustering 
### 1.Segmentation using Age and Spending Score

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np

X1 = df[['Age' , 'Spending Score (1-100)']].values

# k distance plot to estimate/determining what is epsilon
# minPts >= dimensions+1, by rule of thumb setting minPts as 2*dimension

minpoints = 2*2
nn = NearestNeighbors(n_neighbors = minpoints).fit(X1)
distances, idx = nn.kneighbors(X1)
avg_dist = [d.mean() for d in distances]
avg_dist.sort(reverse = True)

# Calculate the difference between consecutive distances
diff = np.diff(avg_dist)

# Find the index of maximum difference
idx = np.argmax(diff)

eps = avg_dist[idx] # Epsilon value
print(eps)

# Draw the horizontal line at the point of maximum increase
plt.plot(list(range(1, len(avg_dist)+1)), avg_dist)
plt.axhline(y=eps, c='r', linestyle='dashed')
plt.title('K Distance Plot')
plt.show()
```

```python
# fitting dbscan to data using parameter estimated from k distance plot
from sklearn.cluster import DBSCAN

db_model = DBSCAN(eps = eps, min_samples = minpoints, n_jobs=-1).fit(X1)
labels = db_model.labels_

# Visualising the clusters
plt.figure(1 , figsize = (15 , 7))
plt.scatter(X1[:,0], X1[:,1], c=labels, cmap='rainbow')
plt.ylabel('Spending Score (1-100)') 
plt.xlabel('Age')
plt.show()

```
